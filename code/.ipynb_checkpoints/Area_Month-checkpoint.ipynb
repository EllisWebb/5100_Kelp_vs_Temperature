```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def plot_monthly_area_temp(df, site_code):

    # Filter by site
    d = df[df["Site Code"] == site_code].copy()

    # Compute average Acres and Temp by month
    monthly_avg = (
        d.groupby("Survey Month")
         .agg(
             Avg_Acres=("Acres", "mean"),
             Avg_Temp=("Temp_t", "mean")
         )
         .reset_index()
    )

    # ---- Side-by-side layout ----
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # LEFT: Boxplot of Acres by month
    sns.boxplot(
        data=d,
        x="Survey Month",
        y="Acres",
        color="lightblue",
        width=0.6,
        ax=axes[0]
    )
    axes[0].set_title(f"Acres by Month — {site_code}")
    axes[0].set_xlabel("Month")
    axes[0].set_ylabel("Acres")
    axes[0].grid(alpha=0.2)

    # RIGHT: Scatterplot of average monthly temp
    axes[1].scatter(
        monthly_avg["Survey Month"],
        monthly_avg["Avg_Temp"],
        color="red",
        s=80
    )
    axes[1].set_title(f"Average Temperature by Month — {site_code}")
    axes[1].set_xlabel("Month")
    axes[1].set_ylabel("Avg Temp (°C)")
    axes[1].grid(alpha=0.2)

    plt.tight_layout()
    plt.show()

    return monthly_avg
```

```python
allbed = pd.read_csv("/Users/carterwebb/Desktop/kelps/data/cleaned_data/corr_dfs/ALL_corrDF.csv")
```

```python
plot_monthly_area_temp(allbed, "EBL")
```

```python
plot_monthly_area_temp(allbed, "SHPT-W")
```

```python
plot_monthly_area_temp(allbed, "SHPT-E")
```
