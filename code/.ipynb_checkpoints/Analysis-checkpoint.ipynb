### Define a fucntion to standaridize the names for the beds

```python


def dataimport(bed_vars, path, name_map):
    for var in bed_vars:
        df = pd.read_csv(path + var + ".csv")
        clean_name = name_map[var]
        globals()[clean_name] = df
        print(f"Loaded {var} → variable: {clean_name}")



```

```python
import urllib.parse

def dataimport(bed_vars, path, name_map):
    for github_filename_key, python_var_name in name_map.items():
        # URL-encode the filename part to handle spaces and other special characters
        filename_to_encode = github_filename_key + ".csv"
        encoded_filename = urllib.parse.quote(filename_to_encode)
        file_url = path + encoded_filename
        try:
            df = pd.read_csv(file_url)
            globals()[python_var_name] = df
            print(f"Loaded {github_filename_key} -> variable: {python_var_name}")
        except Exception as e:
            print(f"Error loading {file_url}: {e}")
```

```python
dataimport(bed_vars, path, name_map)
```
